```python
import pandas as pd
import numpy as np
import thinkplot
import thinkstats2

df2019 = pd.read_csv("ACSST1Y2019.csv")
df2021 = pd.read_csv("ACSST1Y2021.csv")
df = pd.read_csv("raw_data.csv")
df
```

```python
df['log_total_deaths'] = np.log(df['total_deaths']+1)
```

```python
hist = thinkstats2.Hist(df.log_total_deaths,label="log_total_deaths")
thinkplot.Hist(hist)
thinkplot.Config(ylim=[0,100])
thinkplot.Show(xlabel = 'log_total_deaths', ylabel = 'count')
```

```python
hist = thinkstats2.Hist(df.total_deaths,label="total_deaths")
thinkplot.Hist(hist)
thinkplot.Config(xlim=[-100,500])
thinkplot.Show(xlabel = 'total_deaths', ylabel = 'count')
```

```python
hist = thinkstats2.Hist(df.stringency_index,label="stringency_index")
thinkplot.Hist(hist)
thinkplot.Show(xlabel = 'stringency_index', ylabel = 'count')
```

```python
import math
hist = thinkstats2.Hist(df.location,label="location")
thinkplot.Hist(hist)
thinkplot.Show(xlabel = 'location', ylabel = 'count')
```

```python
hist = thinkstats2.Hist(df.gdp_per_capita,label="gdp_per_capita")
thinkplot.Hist(hist)
thinkplot.Show(xlabel = 'gdp_per_capita', ylabel = 'count')
```

```python
hist = thinkstats2.Hist(df.human_development_index,label="human_development_index")
thinkplot.Hist(hist)
thinkplot.Show(xlabel = 'human_development_index', ylabel = 'count')
```

```python
df_numerical = df.select_dtypes(include=[np.number])

# Calculate the Z-score of each value
z_scores = (df_numerical-df_numerical.mean())/df_numerical.std()

# Remove outliers based on the Z-score method
df_numerical = df_numerical[(z_scores < 3).all(axis=1)]
```

```python
df_numerical.nunique()
```

```python
hist = thinkstats2.Hist(df_numerical["total_cases"])
thinkplot.Hist(hist)
thinkplot.Show()
```

```python
# Plot histograms for each column
for column in df_numerical.columns:
    hist = thinkstats2.Hist(df_numerical[column])
    thinkplot.Hist(hist)
    thinkplot.Show()
```
